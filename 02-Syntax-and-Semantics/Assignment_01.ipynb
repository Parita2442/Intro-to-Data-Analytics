# Question 10 & 11

Write a **function** that receives a list of integer numbers and returns a list of numbers that are multiples of 4. Call this function `mult4_filter()`.

**Given the list `L3` below, how many elements does the outcome of `mult4_filter(L3)` have?**

Hint: use `len(mult4_filter(L3))` to get the number of elements.

```python
L3 = [15, 11,  1,  3, 13,  3, 14, 16, 17, 17,  6, 18, 10, 19,  8,  1, 18,
      17, 14,  1,  5,  2, 13,  0,  1, 13, 16,  8,  5, 11, 12,  8, 17, 14,
      10, 18, 17, 16,  3,  7,  8, 15, 18,  7, 10,  5,  7, 16,  6,  5, 24,
      18, 22, 32]
```

```python
# Your answer goes here
def mult4_filter(L):    # define a new function
    # Your code goes here
    return [val for val in L if val%4==0]   # creates a list of values from the input list if it fulfils the divisibilty by 4 condition
```

```python
print(mult4_filter(L3))   # prints the created list after L3 is passed through the function mult4_filter
len(mult4_filter(L3))   # calculates the number of elements in the list created
```
